```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

toys = np.array([[1,1],[1.2,0.9],#bear
          [5,5],[5.2,5.1],#bricks
          [6,6],[6.1,5.9]#cars
])

Kmeans = KMeans(n_clusters = 2, random_state = 42)
Kmeans.fit(toys)
plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,0],marker = "x",s = 100,label = "centroids")

new_data = np.array([[2.0,1.3]])
plt.scatter(new_data[:,0],new_data[:,1],color = 'black',label = "newdata")
plt.scatter(toys[:,0],toys[:,1],c = Kmeans.labels_,label = "ad",cmap = 'rainbow')

prediction = Kmeans.predict(new_data)
print(prediction)

plt.legend()
plt.grid(True)
plt.show()
```
